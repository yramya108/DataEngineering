```python
import pandas as pd
import numpy as np
employees = pd.DataFrame({
    'EmpId' : [101,102,103,104,105,106,107],
    "EmpName" : ['Alice','Bob','Charlie','David','Eve','Frank',None],
    'DeptID':[1,2,1,3,np.nan,2,1]
})
departments = pd.DataFrame({
    'DeptID': [1,2,3,1,1,None,3],
    'DeptName':['HR','IT','Finance','Admin','Sales',None,'Finance']
})
```

```python
import pandas as pd
import numpy as np
customers = pd.DataFrame({
    'CustomerId' : [1,2,3,4,5],
    "CustomerName" : ['Alice','Bob','Bob','David','Eve'],
    'City':['Bangalore','Mumbai','UP','Orissa','AP']
})
orders = pd.DataFrame({
    'OrderId': [101,102,103,104,104],
    'CustomerId':[1,2,3,4,6],
    'Amount' : [200,400,250,100,301]
})
```

```python
pd.merge(customers,orders,on='CustomerId',how='left')
```

```python
pd.merge(customers,orders,on='CustomerId',how='right',indicator=True)
```

```python
pd.merge(customers,orders,on='CustomerId',how='right',indicator=True)
```

```python
pd.merge(customers,orders,on='CustomerId',how='outer',indicator=True)
```
