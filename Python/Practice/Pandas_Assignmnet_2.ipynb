```python
data = {
    "invoice_date": ["2023-04-01", "2023-04-01", "2023-04-02", "2023-04-02", "2023-04-03"],
    "revenue": [100, 150, 200, 250, 300],
}
df1 = pd.DataFrame(data)
```

```python
df1["invoice_date"] = pd.to_datetime(df1["invoice_date"])
```

```python
daily_revenue = df1.groupby("invoice_date")["revenue"].sum().reset_index()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(daily_revenue["invoice_date"], daily_revenue["revenue"], marker="o", linestyle="-", color="blue")
plt.title("Daily Revenue Trends", fontsize=16)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Revenue", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
